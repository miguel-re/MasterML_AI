Calcula la longitud de ambos vectores. 

Encuentra el ángulo que forman los vectores entre sí y genera un nuevo vector que sea ortogonal a cualquiera de los dos primeros. 

```python
# Generación de vectores
x = np.linspace(0,10,100)
vector1 = np.array([1,2])
vector2 = np.array([4,5])

#producto punto
dot_product = np.dot(vector1,vector2)
#Opcion sin libreria np
dot_product_np = vector1[0] * vector2[0] + vector1[1] * vector1[1]

#Distancia de cada vector
dist1 = np.linalg.norm(vector1)
dist2 = np.linalg.norm(vector2)
#Opcion sin libreria np
dist1_np = np.sqrt(vector1[0]**2+vector1[1]**2)
dist2_np = np.sqrt(vector2[0]**2+vector2[1]**2)

#Angulo que forman los vectores
cos_theta = dot_product/(dist1*dist2)
ang_rad = np.arccos(cos_theta)

angulo=np.degrees(ang_rad)

#Vector ortogonal
oct_vect1 = np.array([-vector1[0],vector1[1]])

# Plot
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1 (1, 2)')
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2 (4, 5)')
plt.quiver(0, 0, oct_vect1[0], oct_vect1[1], angles='xy', scale_units='xy', scale=1, color='purple', label='Vector ortogonal')
plt.xlim(-2, 6)
plt.ylim(-1, 7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vectores')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()

plt.text(1, 4, f'Producto punto: {dot_product}', color='black')
plt.text(.35,.5,f'Angulo: {angulo:.2f}', color='black')

plt.text(1,2,f'Distancia 1: {dist1:.2f}', color='blue')
plt.text(4,5,f'Distancia 2: {dist2:.2f}', color='red')

plt.show()
```

## Ejercicio 3

Realiza las siguientes operaciones
- Genera una matriz cuadrada $R^{nxn}$ con $n>3$ y calcula su determinante y su traza
- Calcula el rango de la matriz generada y los valores de los coeficientes de su polinomio característicos
- Encuentra las raíces polinomio característico de la matriz generadalotlib. 

```python
#Matriz
matrix = np.random.randint(1, 11, size=(4, 4))
#Determinante y traza
determinante = np.linalg.det(matrix)
traza = np.trace(matrix)
#Rango de la matriz
rango = np.linalg.matrix_rank(matrix)
#Valores de coeficientes
coeficientes = np.poly(matrix)
#Valores propios [Raices del polinomio característico]
valores_propios = np.linalg.eigvals(matrix)

print(f'Matriz: \n{matrix}')
print(f'Determinante: {determinante}')
print(f'Traza: {traza}')
print(f'Rango: {rango}')
print(f'Coeficientes: {coeficientes}')
print(f'Valores propios: {valores_propios}')
```

## Ejercicio 4
Calcula los autovalores y autovectores de la siguiente matriz [5,2],[2,2] realiza su autodescomposición y grafica los resultados en un plano Matplotlib

```python
#Matriz
A = np.array([[5, 2],
              [2, 2]])

# Calcular los autovalores y autovectores
autovalores, autovectores = np.linalg.eig(A)

# Mostrar autovalores y autovectores
print("Autovalores:\n", autovalores)
print("Autovectores:\n", autovectores)

# Autodescomposición: A = PDP^(-1)
P = autovectores
D = np.diag(autovalores)

# Graficar
origin = np.array([[0, 0], [0, 0]])  # origen para los autovectores
plt.quiver(*origin, autovectores[0], autovectores[1], color=['r', 'b'], scale=5, label='Autovectores')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.grid()
plt.title('Autovectores de la matriz')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
