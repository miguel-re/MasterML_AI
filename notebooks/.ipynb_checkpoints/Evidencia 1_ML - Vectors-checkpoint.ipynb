## Pregunta 1

Resuelve las siguientes ecuaciones lineales
$$ a. 8x + 5y + 3z - 6w = 41$$
$$ 15x - 3y + 4z + 7w = 13$$
$$ x + 6y - 2z + w = 12$$
$$ -2x + y + 7z - 3w = 0$$

```python
#Coefs
A = np.array([[8,5,3,-6],
              [15,-3,4,7],
              [1,6,-2,1],
              [-2,1,7,-3]])
#Terminos
B = np.array([41,13,12,0])
```

## Pregunta 2

Dado el vector base $\begin{bmatrix}
2 \\
5
\end{bmatrix} \in R^{2}$, que genera el subespacio unidimensional $\mu \subseteq R^{2}$. Encuentra el valor de la coordenada $\lambda$, la proyección $\pi_{\mu}(x) \in \mu$ y la matriz $P_{\pi}$ para proyectar sobre el subespacio $\mu$ el vector $\begin{bmatrix}
-1 \\
4
\end{bmatrix} \in R^{2}$

Para resolver este problema, utilizaremos la fórmula: 
$$\lambda = \frac{b^{T}x}{b^{T}b}$$
donde $x = \begin{bmatrix}
-1 \\
4
\end{bmatrix}$
Se calcula primero $b^{T}x$ y $b^{t}b$: 
$$b^{T}x = \begin{bmatrix}
2 & 5
\end{bmatrix} \begin{bmatrix}
-1 \\
4
\end{bmatrix} = 2(-1)+5(4) = -2 + 20 = 18$$

$$b^{T}b = \begin{bmatrix}
2 & 5
\end{bmatrix} \begin{bmatrix}
2 \\
5
\end{bmatrix} = 2^{2}+5^{2} = 4 + 25 = 29$$

Por lo tanto, $\lambda$ es:

$$\lambda = \frac{18}{29}$$

El paso 2 es el cálculo de la proyección $\pi_{\mu}(x)$

La proyección del vector $x$ sobre el subespacio $\mu$ es:

$$\lambda_{\mu}(x) = \lambda b$$

Sustituyendo el valor de $\lambda$:

$$\pi_{\mu}(x) = \frac{18}{29} \begin{bmatrix}
2 \\
5
\end{bmatrix} = \begin{bmatrix}
\frac{36}{29} \\
\frac{90}{29}
\end{bmatrix} $$

Por último, se construye la matriz de proyección $P_{\pi}$

$$P_{\pi} = \frac{bb^{T}}{b^{T}b}$$

Calculando $bb^{T}$:
$$bb^{T} = \begin{bmatrix}
2 & 5
\end{bmatrix} \begin{bmatrix}
2 \\
5
\end{bmatrix} = \begin{bmatrix}
4 & 10 \\
10 & 25
\end{bmatrix}$$

Entonces, la matriz de proyección es:

$$P_{\pi} = \frac{1}{29} \begin{bmatrix}
4 & 10 \\
10 & 25
\end{bmatrix} = \begin{bmatrix}
\frac{4}{29} & \frac{10}{29} \\
\frac{10}{29} & \frac{25}{29}
\end{bmatrix}$$


Obteniendo:

- Valor de $\lambda$: $\frac{18}{29}$
- Proyección de $\pi_{\mu}(x)$: $\begin{bmatrix}
\frac{36}{29} \\
\frac{90}{29}
\end{bmatrix}$
- Matriz de proyeccion $P_{\pi}$: $\begin{bmatrix}
\frac{4}{29} & \frac{10}{29} \\
\frac{10}{29} & \frac{25}{29}
\end{bmatrix}$

```python
#Utilizando funciones de python
b = np.array([2,5])

x = np.array([-1,4])

#Lambda = b^{t}x/b^{t}b
lambda_val = np.dot(b,x)/np.dot(b,b)

#Proyeccion = lambda * b
proyection = lambda_val * b

#Matriz de proyeccion = bb^{t} / b^{t}b
P_pi = np.outer(b,b)/np.dot(b,b)

#Gráficamente
plt.figure(figsize = (8,6))

plt.quiver(0,0,x[0], x[1], angles = 'xy', scale_units = 'xy', scale=1, color='r', label='Vector x')
plt.quiver(0,0,b[0], b[1], angles = 'xy', scale_units = 'xy', scale=1, color='b', label='Vector base')
plt.quiver(0,0,proyection[0], proyection[1], angles = 'xy', scale_units = 'xy', scale=1, color='g', label='Projection')

plt.xlim(-2,3)
plt.ylim(-1,6)
plt.axhline(0,color='black', linewidth=.5, ls='--')
plt.axvline(0,color='black', linewidth=.5, ls='--')
plt.grid()
plt.title('Proyeccion de un vector en un subespacio')
plt.legend()
plt.show()
```

```python
print(f'La matriz es: \n{P_pi}')
```

```python
eigenval, eigenvec = np.linalg.eig(P_pi)
```

Recordemos que los eigevalores indican la escala de la transformación aplicada sobre la matriz, mientras que los eigevectores indican las direcciones invariantes bajo esta transformación.

```python
prueba1
```

```python
#Prueba 1
df_temp = df[['EDAD', 'ID_PERSONA', 'ENT', 'SEXO', 'MUN', 'ESCOLARI']]
df_temp = df_temp[(df_temp['EDAD'] >= 5) & (df_temp['EDAD']<=90)]

prueba1 = df_temp['EDAD'].value_counts().sort_index()

plt.figure(figsize=(15,6))
prueba1.plot(kind='bar', color='skyblue')
plt.title('Edades')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()

print(f'- Media: {np.mean(prueba1)} \n- std: {np.std(prueba1)}')
```

```python
df_temp = df[df['ESCOLARI'].isin([3,4,5,6])]
df_temp = df_temp[(df_temp['EDAD'] >= 14) & (df_temp['EDAD'] <= 88)]
prueba2 = df_temp.groupby('EDAD', as_index=False).count()[['EDAD', 'ID_PERSONA']]

plt.figure(figsize=(12,6))
plt.bar(prueba2['EDAD'], prueba2['ID_PERSONA'], color='skyblue')
plt.title('Distribución por edades de personas con estudios de secundaria mínimos')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(prueba2['EDAD'],rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()

print(f'- Media: {np.mean(prueba2['ID_PERSONA'])} \n- std: {np.std(prueba2['ID_PERSONA'])}')
```

```python
df_temp = df[(df['EDAD'] >= 12) & (df['EDAD'] <= 88)]
df_temp = df_temp[df_temp['SITUA_CONYUGAL'].isin([5,6,7])]
prueba3 = df_temp.groupby('EDAD', as_index=False).count()[['EDAD', 'ID_PERSONA']]

plt.figure(figsize=(12,6))
plt.bar(prueba3['EDAD'], prueba3['ID_PERSONA'], color='skyblue')
plt.title('Distribución por edades de personas casadas')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(prueba3['EDAD'],rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()

print(f'- Media: {np.mean(prueba3['ID_PERSONA'])} \n- std: {np.std(prueba3['ID_PERSONA'])}')
```

La mejor forma para demostrar la veracidad de las gráficas anteriores con una distribución normal, es mediante una prueba de normalidad. En este caso, usaremos la prueba de Kolmogorov-Smirnov.

Dicha prueba genera 2 resultados principales, por un lado es el estadístico de K-S, el cual representa la diferencia entre la función de distribución empírica de la muestra. De igual manera, genera el valor p, el cual indica la probabilidad de observar un estadístico K-S tan extremo como el calculado.

Se propone una hipotesis nula, la cual propone un nivel de significancia de .05 y en caso que el p-value sea menor a ese valor, se rechaza la hipotesis nula. Por ende hay suficiente evidencia para afirmar que la muestra proviene de la distribución específica.

```python
#Prueba 1 - KS
from scipy import stats

ks_test = stats.kstest(prueba1,'norm', args=(np.mean(prueba1), np.std(prueba1)))

print('K-S Test:', ks_test)
```

```python
ks_test = stats.kstest(prueba2['ID_PERSONA'],'norm', args=(np.mean(prueba2['ID_PERSONA']), np.std(prueba2['ID_PERSONA'])))

print('K-S Test:', ks_test)
```

```python
ks_test = stats.kstest(prueba3['ID_PERSONA'],'norm', args=(np.mean(prueba3['ID_PERSONA']), np.std(prueba3['ID_PERSONA'])))

print('K-S Test:', ks_test)
```

Tras realizar las 3 pruebas, tenemos 3 distintos resultados, donde hay *p-values* con 3 posibles resultados, uno por debajo del valor propuesto de la hipotesis nula (prueba 3), y 2 por encima de la hipotesis nula (prueba 1 y 2). Se puede evidenciar que la más acercada a una distribución normal es la prueba 3.

Aún así realizaré la prueba del gráfico de la normal

```python
#Prueba 1
mu = pd.DataFrame(prueba1.index)['EDAD'].mean()
std = pd.DataFrame(prueba1.index)['EDAD'].std()

x = np.linspace(prueba1.index.min(), prueba1.index.max(), 1400)
y = stats.norm.pdf(x, mu, std) * prueba1.sum() * (x[20] - x[0])

plt.figure(figsize=(15,6))
plt.plot(x,y,color='red')
prueba1.plot(kind='bar', color='skyblue')
plt.title('Edades')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()
```

```python
#Prueba 2
mu = prueba2['EDAD'].mean()
std = prueba2['EDAD'].std()

x = np.linspace(prueba2['EDAD'].min(), prueba2['EDAD'].max(), 1400)
y = stats.norm.pdf(x, mu, std) * prueba2['ID_PERSONA'].sum() * (x[20] - x[0])

plt.figure(figsize=(12,6))
plt.plot(x,y,color='red')
plt.bar(prueba2['EDAD'], prueba2['ID_PERSONA'], color='skyblue')
plt.title('Distribución por edades de personas con estudios de secundaria mínimos')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(prueba2['EDAD'],rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()
```

```python
#Prueba 2
mu = prueba3['EDAD'].mean()
std = prueba3['EDAD'].std()

x = np.linspace(prueba3['EDAD'].min()-1, prueba3['EDAD'].max()+1, 900)
y = stats.norm.pdf(x, mu, std) * prueba3['ID_PERSONA'].sum() * (x[25] - x[10])

plt.figure(figsize=(12,6))
plt.plot(x,y,color='red', linewidth=2)
plt.bar(prueba3['EDAD'], prueba3['ID_PERSONA'], color='skyblue')
plt.title('Distribución por edades de personas con estudios de secundaria mínimos')
plt.xlabel('EDAD')
plt.ylabel('Conteo')
plt.xticks(prueba3['EDAD'],rotation=90)
plt.grid(axis='y', linestyle='--', alpha=.7)
plt.show()
```

Tal como se observa en los resultados, la prueba 3 es la más acercada a una distribución normal
